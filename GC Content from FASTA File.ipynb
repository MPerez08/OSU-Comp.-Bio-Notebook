```python
import io

def base_composition(seq, query_base):
    base_counter = 0
    seq_len = len(seq)
    
    for index in range(0, seq_len):
        seq_base = seq[index]
        if seq_base == query_base:
            base_counter = base_counter + 1
    
    return base_counter
## Given a DNA (A,C,T,G) sequence string, returns the GC-content as a gloat

def gc_content(seq):
    g_cont = base_composition(seq, "G") #1st function embedded to this function
    c_cont = base_composition(seq, "C")
    seq_len = len(seq)
    gc = (g_cont + c_cont)/float(seq_len)*100
    return "GC content is: " + str(gc)

##Open file, and loop over lines
import io
fhandle = io.open("id_seqs_ch15.txt")

for line in fhandle:
    line_stripped = line.strip()
    linelist = line.split("\t")
    id = linelist[0]
    sequence = linelist[1]
    seqgc = gc_content(sequence)
    print(id + "\t" + str(seqgc))

'''COnceptual Understanding of above code'''
''' 4 control flows applied above are: 
1) Conditional statements: evaluates a parameter for certain criteria
E.g) if seq_base == query_base

2) "For" Loops: iterates a parameter to assess certain criteria
Eg for index in range(0, seq_len)

3) Range Function: will give the specific length for a parameter to be assessed
Eg) range(0, seq_len) 
```
